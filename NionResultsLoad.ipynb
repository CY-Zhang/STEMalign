```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
path = '../../NionData/102121/'
idx = 3
Y = np.squeeze(np.load(path + 'FirstOrder_45mrad_500ms_' + str(idx) + '_Y.npy'))
Ronch = np.load(path + 'FirstOrder_45mrad_500ms_' + str(idx) + '_Ronchigram.npy')
plt.imshow(Ronch[-1,:,:])
plt.axis('off')
print(Y)
print(Ronch.shape)
```

```python
path = '../../NionData/102121/'
fig, ax = plt.subplots(1, 1, figsize = [6, 3])
n = 5
for idx in range(n):
    Y = np.squeeze(np.load(path + 'FirstOrder_45mrad_500ms_' + str(idx) + '_Y.npy'))
    X = np.load(path + 'FirstOrder_45mrad_500ms_' + str(idx) + '_X.npy')
    Ronch = np.load(path + 'FirstOrder_45mrad_500ms_' + str(idx) + '_Ronchigram.npy')
    ax.plot(1-Y, c = 'C0', alpha = 1.0)
ax.set_ylabel('CNN prediction', fontsize = 16)
ax.set_xlabel('Iteration', fontsize = 16)
ax.tick_params(axis='both', labelsize=16)
ax.axvline(x=10, ls = '--', color = 'black')
```

```python
path = '../../NionData/102121/'
fig, ax = plt.subplots(3, 1, figsize = [6, 10])
n = 5
for idx in range(n):
    Y = np.squeeze(np.load(path + 'FirstOrder_45mrad_500ms_' + str(idx) + '_Y.npy'))
    X = np.load(path + 'FirstOrder_45mrad_500ms_' + str(idx) + '_X.npy')
    Ronch = np.load(path + 'FirstOrder_45mrad_500ms_' + str(idx) + '_Ronchigram.npy')
    ax[0].plot((X[:,0] - 0.5) * 2, c = 'C0', alpha = 1.0)
    ax[1].plot((X[:,1] - 0.5) * 0.15, c = 'C1', alpha = 1.0)
    ax[2].plot((X[:,2] - 0.5) * 0.15, c = 'C2', alpha = 1.0)
    
ax[0].set_ylabel('C10 ($\mu$m)', fontsize = 16)
ax[1].set_ylabel('C12.x ($\mu$m)', fontsize = 16)
ax[2].set_ylabel('C12.y ($\mu$m)', fontsize = 16)
ax[2].set_xlabel('Iteration', fontsize = 16)
ax[0].tick_params(axis='both', labelsize=16)
ax[1].tick_params(axis='both', labelsize=16)
ax[2].tick_params(axis='both', labelsize=16)
ax[0].axvline(x=10, ls = '--', color = 'black')
ax[1].axvline(x=10, ls = '--', color = 'black')
ax[2].axvline(x=10, ls = '--', color = 'black')
```
